```python
from selenium.webdriver import Firefox
from selenium.webdriver.firefox.options import Options
```

```python
path = 'geckodriver.exe'
opts = Options()
opts.headless = True
browser = Firefox(options=opts)
```

```python
def process_box(box_class):
    '''
        Construct and return dict corresponding to box with class name box_class
    '''
    review_boxes = browser.find_elements_by_css_selector(box_class)

    res_list = []
    for review_box in review_boxes:
        cur_dict = dict()
        
        course_num = review_box.find_element_by_css_selector('.course_number')
        cur_dict['number'] = course_num.text

        prof = review_box.find_element_by_xpath('.//a')
        cur_dict['professor'] = prof.text

        review_content = review_box.find_element_by_css_selector('.review_content')
        cur_dict['review'] = review_content.text

        res_list.append(cur_dict)
    return res_list
    
def process_pages(browser, links):
    res = []
    
    for link in links:
        browser.get(link)
        
        try:
            res += process_box('.review.box.old')
        
        except:
            try:
                res += process_box('.review.box')
            except Exception as ex:
                print(ex)
                
    return res

browser.get('http://culpa.info/departments/7')
course_box = browser.find_element_by_css_selector('.box.courses')
links = [link.get_attribute('href') for link in course_box.find_elements_by_xpath('.//a')]
reviews = process_pages(browser, links)
```

```python
keys = sorted([review['number'] for review in reviews])
```

```python
refined_dict=[]
for i in reviews:
    for k,v in i.items():
        if len(v) != 12:
            break;
        if int(v[7]) < 4:
            break;
        
        new_v = v[6]+ v[1:5] + v[7:-1]
        i[k] = new_v
        refined_dict.append(i)
        break
```

```python
keys = sorted([review['number'] for review in refined_dict])
keys
```

```python
import pandas as pd

df = pd.DataFrame(refined_dict)

df.to_csv('review.csv', index = False)
```
